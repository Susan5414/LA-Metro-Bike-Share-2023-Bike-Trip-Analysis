### Combine four quarters data together

```python
# Concatenate data frame
combined_q1q2 = pd.concat([q1, q2], axis=0)
combined_q3q4 = pd.concat([q3, q4], axis=0)
```

```python
df_all = pd.concat([combined_q1q2, combined_q3q4], axis=0) # Combine all 4 quarters
```

```python
df_all
```

Total rows and columns are correct.

### Data Cleaning & Consistency Check

```python
# Check mixed data type
for col in df_all.columns.tolist():
  weird = (df_all[[col]].map(type) != df_all[[col]].iloc[0].map(type)).any(axis = 1)
  if len (df_all[weird]) > 0:
    print (col)
```

No mixed data type.

```python
# Check missing value
df_all.isnull().sum()
```

No missing vlaue.

```python
# Check duplications
df_dups = df_all[df_all.duplicated()]
```

```python
df_dups
```

No duplicates.
